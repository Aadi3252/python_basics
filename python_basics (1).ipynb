Q1 Explain the key features of Python that make it a popular choice for programming.
Python is widely used due to the following key features:

Easy to Learn and Use:
Python has a simple and readable syntax that resembles English, making it beginner-friendly.

Interpreted Language:
Python executes code line by line, which makes debugging easier and faster.

Dynamically Typed:
You don’t need to declare data types; Python automatically detects the type during runtime.

High-Level Language:
Python handles complex tasks like memory management automatically, allowing programmers to focus on logic.

Extensive Standard Library:
Python comes with a large collection of built-in modules and functions, supporting tasks like file handling, regular expressions, and web development.

Portability:
Python code can run on different platforms (Windows, macOS, Linux) without modification.

Object-Oriented and Functional:
Python supports multiple programming paradigms, including object-oriented, procedural, and functional programming.

Large Community Support:
A strong community helps with resources, libraries, and frameworks, making development faster and easier.

Third-party Modules:
Python has a rich ecosystem of external libraries (like NumPy, Pandas, Django) that extend its capabilities.

Q3  Compare and contrast mutable and immutable objects in Python with examples

Mutable objects:
Objects that can be changed after they are created. Their content (value) can be modified without changing their identity (memory location).

Immutable objects:
Objects that cannot be changed once they are created. Any operation that seems to modify them actually creates a new object.

Q4   Discuss the different types of operators in Python and provide examples of how they are used

1. Arithmetic Operators

a = 10
b = 3
print(a + b)   # Output: 13
print(a - b)   # Output: 7
print(a * b)   # Output: 30
print(a / b)   # Output: 3.33
print(a % b)   # Output: 1
print(a // b)  # Output: 3 (floor division)
print(a ** b)  # Output: 1000 (exponentiation)


2. Assignment Operators

x = 5          # assigns 5 to x
x += 2         # x becomes 7
x *= 3         # x becomes 21
x -= 1         # x becomes 20

3. Comparison Operators

a = 5
b = 3
print(a == b)  # False
print(a != b)  # True
print(a > b)   # True
print(a < b)   # False
print(a >= b)  # True
print(a <= b)  # False

4. Logical Operators

a = 5
b = 3
print(a > 2 and b < 5)  # True
print(a < 2 or b < 5)   # True
print(not(a > b))       # False



5. Bitwise Operators

a = 5      # binary: 0101
b = 3      # binary: 0011
print(a & b)   # 1
print(a | b)   # 7
print(a ^ b)   # 6
print(~a)      # -6
print(a << 1)  # 10 (left shift)
print(a >> 1)  # 2  (right shift)


6. Membership Operators

print('a' in 'apple')         # True
print('x' not in 'banana')    # True

7. Identity Operators

x = [1, 2, 3]
y = x
z = [1, 2, 3]

print(x is y)       # True (same object)
print(x is z)       # False (same content, different object)
print(x is not z)   # True

Q5 Explain the concept of type casting in Python with examples

Type casting in Python refers to converting one data type into another. It is useful when you want to perform operations involving different data types or ensure that values match a required format.

There are two types of type casting:

1. Implicit Type Casting
This is done automatically by Python when it converts one data type to another during an operation.

a = 5       # int
b = 2.0     # float
c = a + b   # Python automatically converts 'a' to float
print(c)    # Output: 7.0 (float)

2. Explicit Type Casting
This is done manually by the programmer using built-in functions like int(), float(), str(), list(), etc.

x = 7.8
y = int(x)
print(y)    # Output: 7
